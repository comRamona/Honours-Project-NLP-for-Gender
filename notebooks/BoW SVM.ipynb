```python
from os import listdir
from os.path import isfile, join
from os import environ
import logging
import os
import pandas as pd
import re
from metadata import Gender
from metadata.metadata import ACL_metadata
import _pickle as pkl
from _name_classification.nametools import process_str as process_str
from _storage.storage import FileDir
```

```python
acl = ACL_metadata()
fd = FileDir()
files = acl.modeling_files
df = acl.modeling_df
```

```python
tok_ids = fd.load_pickle("doc10_ids")
```

```python
df
```

```python
from collections import defaultdict
single_fem = defaultdict(int)
single_male = defaultdict(int)
for i, file in enumerate(acl.modeling_files):
    gender = df.loc[acl.get_id(file)]["genders"]
    if len(gender) == 1:
        authors= df.loc[acl.get_id(file)]["authors"]
        if gender[0] == Gender.female:
            single_fem[authors[0]] += 1
        elif gender[0] == Gender.male:
            single_male[authors[0]] += 1
```

```python
len(single_fem)
```

```python
df.loc[tok_ids[0]]
```

```python
males = []
females = []
for i, file in enumerate(acl.modeling_files):
    authors = df.loc[acl.get_id(file)]["authors"]
    if len(authors) == 1:
        a = authors[0]
        if single_fem[a] != 0:
            females.append(acl.get_id(file))
            single_fem[a] = 0
        elif single_male[a] != 0:
            males.append(acl.get_id(file))
            single_male[a] = 0
            
    
```

```python
import random
len(females)
```

```python
sample_size = len(females)
sample_index = sorted(random.sample(range(len(males)), sample_size))
```

```python
men_sample = [males[j] for j in sample_index]
fem_sample = females
x = men_sample + fem_sample
labels = [0] * len(men_sample) + [1] * len(fem_sample)
```

```python
from gensim.corpora import MmCorpus, Dictionary
corpus = MmCorpus(os.path.join(fd.models, "acl_bow10.mm"))
docs = fd.load_pickle("docs10" )

dictionary = Dictionary(docs)

# max_freq = 0.4
# min_wordcount = 5
# max_number = 20000
dictionary.filter_extremes(keep_n = 10000)
_ = dictionary[0]
```

```python
sample_docs = []
for i in x:
    sample_docs.append(docs[tok_ids.index(i)])
```

```python
corpus = [dictionary.doc2bow(doc) for doc in sample_docs]
```

```python
import gensim
X = gensim.matutils.corpus2csc(corpus)
```

```python
transp = X.T
```

```python
from sklearn.cross_validation import train_test_split
a_train, a_test, b_train, b_test = train_test_split(transp, labels, test_size=0.2, random_state=42)
```

```python
from sklearn import svm
clf = svm.LinearSVC()
#clf = svm.SVC()
clf.fit(a_train, b_train)
```

```python
len(labels)
```

```python
preds = clf.predict(a_test)
```
