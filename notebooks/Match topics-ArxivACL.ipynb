```python
from gensim.models.ldamodel import LdaModel
from gensim.models.coherencemodel import CoherenceModel
from gensim.corpora import MmCorpus
from gensim.corpora import Dictionary
from collections import Counter, defaultdict
from numpy.random import seed
from _storage.storage import FileDir
from os.path import join
import re
from metadata.metadata import ACL_metadata
from _topic_modeling.lda_loader import Loader, ArxivLoader
from _storage.storage import FileDir
from os.path import join
import _pickle as pkl
from metadata import Gender
import numpy as np
import logging
import gensim 

seed(1)
logging.basicConfig(format='%(levelname)s : %(message)s', level=logging.INFO)
logging.root.level = logging.INFO

from scipy.stats import entropy

fd = FileDir()
loader1 = Loader(100)
lda1 = loader1.model
topics1 = lda1.get_topics()
topic_corresp = loader1.topic_corresp
topic_corresp[15] = "StatisticalParsing"
        
n_t = 50
loader2 = ArxivLoader(n_t)
lda2 = loader2.model
topics2 = lda2.get_topics()
```

```python
arxiv_d = loader2.dic.id2token
acl_d = loader1.dic.token2id
corresp = defaultdict()
for t in range(n_t):
    v2 = topics2[t]
    score = 10**6
    best_name = ""
    for t1 in range(100):
        v1 = np.array(np.zeros(len(v2)))
        for i in range(len(v1)):
            vv = acl_d.get(arxiv_d[i],-1)
            if vv != -1:
                v1[i] = topics1[t1][vv]     
        result = entropy(v1,v2)
        if(result < score):
            best_name = topic_corresp[t1]
            score = result
    corresp[i] = best_name
    x = lda2.show_topic(t, 10)
    words = ""
    for w, sc in x:
        words += (w + " ")
    print(i, best_name, score, words)
```

```python
fd.save_pickle(corresp, "topic_suggestion50")
```
